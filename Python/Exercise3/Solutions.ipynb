Write a Python program to create a 1x2 grid of subplots. The first subplot should show a simple
line plot with data [1, 2, 3, 4] and [10, 20, 30, 40]. The second subplot should display a bar plot
with the same x-axis values and corresponding y-axis values [5, 10, 15, 20]

```python
x_values = [1, 2, 3, 4]
line_values = [10, 20, 30, 40]
bar_values = [5, 10, 15, 20]

fig, axes = plt.subplots(1, 2, figsize=(10,5), sharex = True)

axes[0].plot(x_values, line_values)
axes[0].set_title("Line Plot")
axes[0].set_ylabel("Y axis")
axes[0].set_xlabel("X axis")

axes[1].bar(x_values, bar_values)
axes[1].set_title("Bar Plot")
axes[1].set_ylabel("Y axis")
axes[1].set_xlabel("X axis")


plt.show()
```

Create a Python program that defines a custom function to calculate the factorial of a number.
The function should take one integer as an argument and return its factorial.

```python
def factorial(n):
  if n == 1:
    return 1
  elif n == 0:
    return 1
  return n * factorial(n-1)

number = 5
print(f"Factorial of {number} is: {factorial(number)}")
```

Write a program to solve the below equations

 Write a program to integrate the function f(x)=x^2 from 0 to 1.

```python
def function_to_integrate(x):
  return x ** 2

lower_limit = 0
upper_limit = 1

result, error = quad(function_to_integrate, lower_limit, upper_limit)
print(f"Integral of x^2 from {lower_limit} to {upper_limit}: {result:.2f} with error: {error:.2f}")
```

Write a program to find the inverse by inv() function and determinant det() function.

```python
matrix = [[1,2,3],
          [4,6,9],
          [-4,3,8]]

determinant = np.linalg.det(matrix)
print("Determinant:", determinant)

if determinant == 0:
  print("Inverse doesnt exist")
else:
  inverse = np.linalg.inv(matrix)
  print("Inverse:", inverse)
```

Write a program to Perform Arithmetic Operations on Multi-Dimensional
Arrays.

```python
def add_matrices(matrix1, matrix2):
  if matrix1.shape != matrix2.shape:
    print("Matrices should be of same shape")
    return None
  return matrix1 + matrix2

def subtract_matrices(matrix1, matrix2):
  if matrix1.shape != matrix2.shape:
    print("Matrices should be of same shape")
    return None
  return matrix1 - matrix2

def multiply_matrices(matrix1, matrix2):
  if matrix1.shape[1] != matrix2.shape[0]:
    print("Matrix sizes not compatible for multiplication")
    return None
  return np.dot(matrix1, matrix2)

matrix1 = np.array([[1, 2, 3],
                   [4, 5, 6],
                    [7, 6, -3]])
matrix2 = np.array([[7, 8, 9],
                   [10, 11, 12], [1, -4, 6]])

print(f"Add {matrix1} + {matrix2} = {add_matrices(matrix1, matrix2)}")
print(f"Subtract {matrix1} - {matrix2} = {subtract_matrices(matrix1, matrix2)}")
print(f"Multiply {matrix1} * {matrix2} = {multiply_matrices(matrix1, matrix2)}")

```

Create a python program to check the missing values in a dataset.

```python
data = {
        'Name': ['Alice', 'Bob', np.nan, 'David', 'Eve'],
        'Age': [25, 30, 22, np.nan, 28],
        'City': ['New York', 'London', 'Paris', 'New York', np.nan],
        'Salary': [50000, 60000, 45000, 70000, 55000]
    }
df = pd.DataFrame(data) # create dataframe

missing_values_count = df.isna().sum().sum()
print(f"Missing values count: {missing_values_count}")
```

Write a python program to group and aggregate the data.

```python
data = {
    'Category': ['Fruits', 'Fruits', 'Vegetables', 'Dairy', 'Fruits', 'Vegetables', 'Dairy'],
    'Item': ['Apple', 'Banana', 'Carrot', 'Milk', 'Orange', 'Spinach', 'Cheese'],
    'Quantity': [10, 15, 20, 5, 12, 8, 3],
    'Price_per_Unit': [1.5, 0.75, 0.5, 2.0, 1.2, 0.8, 4.0]
}
df = pd.DataFrame(data)

total_quantity_by_category = df.groupby('Category')['Quantity'].sum()
print(total_quantity_by_category)

total_agg_quantity = df.groupby('Category')['Quantity'].agg(['sum', 'mean', 'max'])
print(total_agg_quantity)
```
